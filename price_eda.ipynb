## Price Data EDA

This notebook includes the perform of EDA on price of Germany - Austria - Luxembourg bidding zone

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
import os
path = '/Users/aaronpapa/Desktop/Fall 2022/Fall2022CapstoneProject/data/'
path
```

```python
# read weather data
price_2016 = pd.read_csv(path+"Day-ahead Prices_2016.csv")
```

```python
price_2017 = pd.read_csv(path+"Day-ahead Prices_2017.csv")
```

```python
price_2016.iloc[:, :2]
```

```python
price_2017['Time'] = pd.to_datetime(price_2017['MTU (CET/CEST)'].str.slice(0, 16), format='%d.%m.%Y %H:%M')
```

```python
price_2016['Time'] = pd.to_datetime(price_2016['MTU (CET/CEST)'].str.slice(0, 16), format='%d.%m.%Y %H:%M')
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(price_2016['Time'], price_2016['Day-ahead Price [EUR/MWh]']);
ax.set_title("Energy Price in Germany - Austria - Luxembourg bidding zone in 2016",)
ax.set_xlabel("Time per hour")
ax.set_ylabel("Price [EUR/MWh]")
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(price_2017['Time'], price_2017['Day-ahead Price [EUR/MWh]']);
ax.set_title("Energy Price in Germany - Austria - Luxembourg bidding zone in 2017",)
ax.set_xlabel("Time per hour")
ax.set_ylabel("Price [EUR/MWh]")
```

The energy price in germany has a clear seasonal pattern shared by both year 2016 and 2017.

The energy price is stable for most of the days through a year around 40 to 50 Euro per kwh except winter.

Seasonal influence are shown as more fluntuations in winter where several peak prices happen.

There are several date with negative price representing the energy oversupply occurs.


The final information we could get are:

    Price is less stable in winter. 
    Energy oversupply occurs in any season but only keeps for 1 or a few days.
